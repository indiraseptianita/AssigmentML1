```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('/content/drive/MyDrive/mesin learning/datamesinlearning1.xlsx')
df.head
```

```python
df.head()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
  
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype('str') 
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
df.head
```

```python
basket = (df[df['Country'] =="France"] .groupby(['InvoiceNo', 'Description'])['Quantity'] .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
 
```

```python
#menghapus kolom postage
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
# Membuat item 
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
frequent_itemsets.head()
```

```python
# Create the asosiasi rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
```
